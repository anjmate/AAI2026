Original Summary Prompt:

Summarize the following article in exactly 5 sentences.
Keep tone neutral and professional.

Self-Reflection Prompt

Evaluate the summary using these criteria:

1. Accuracy – Does it reflect all major ideas?

2. Clarity – Is it understandable for a college-level audience?

3. Completeness – Does it cover both benefits and risks?

4. Conciseness – Is it exactly 5 sentences?

5. Tone – Is it neutral and professional?

Provide:

* A brief critique under each category.

* Then provide a revised improved summary.

* Keep the improved summary under 120 words.

```python
print("ORIGINAL ARTICLE:\n")

article_text = """
Artificial intelligence is transforming industries by automating processes,
improving decision-making, and enhancing customer experiences.
However, it also raises concerns about job displacement,
algorithmic bias, and ethical responsibility.
Organizations must balance innovation with regulation
to ensure responsible and equitable AI deployment.
"""

print(article_text)


print("\n\nORIGINAL SUMMARY:\n")

# Intentionally weaker summary for improvement demonstration
original_summary = (
    "Artificial intelligence is changing industries by automating tasks and improving decisions. "
    "It helps businesses become more efficient and improve customer experiences. "
    "However, AI may cause job loss. "
    "There are also ethical concerns. "
    "AI has both advantages and disadvantages."
)

print(original_summary)

print("\nSELF-CRITIQUE:\n")

print("1. Accuracy:")
print("- The summary mentions automation and job loss but does not clearly explain algorithmic bias or regulation.")

print("\n2. Clarity:")
print("- The language is simple but overly general and lacks specific detail.")

print("\n3. Completeness:")
print("- It mentions benefits and risks but does not fully describe ethical responsibility or regulation.")

print("\n4. Conciseness:")
print("- It is exactly 5 sentences, which satisfies the requirement.")

print("\n5. Tone:")
print("- Tone is neutral but somewhat simplistic and lacks academic depth.")

print("\nIMPROVED SUMMARY:\n")

improved_summary = (
    "Artificial intelligence is transforming industries by automating processes, enhancing decision-making, "
    "and improving customer experiences. Businesses benefit from greater efficiency and innovation through AI adoption. "
    "However, its expansion raises serious concerns about job displacement, algorithmic bias, and ethical accountability. "
    "Responsible implementation requires balancing technological advancement with regulation and oversight. "
    "Ultimately, AI offers significant opportunities but must be managed carefully to ensure equitable outcomes."
)

print(improved_summary)

```

The original summary lacked depth, specificity, and discussion of regulation and algorithmic bias.

After applying explicit evaluation criteria (accuracy, clarity, completeness, conciseness, tone), the revised summary:

* Includes algorithmic bias and regulation

* Uses more precise academic language

* Maintains exactly five sentences

* Stays under 120 words

* Improves clarity and completeness

This demonstrates a full critique-and-revise cycle aligned with prompt engineering best practices.
