```python
# -----------------------------
# STEP 1: Issue Classification Prompt
# -----------------------------

classification_prompt = """
You are a customer support classification system.

Classify the customer issue into ONE of these categories:
- Billing
- Technical Support
- Account Management
- General Inquiry

Return ONLY the category name.
"""

# -----------------------------
# STEP 2: Urgency Detection Prompt
# -----------------------------

urgency_prompt = """
Determine urgency level:
High = legal threats, fraud, cancellation threats
Medium = strong dissatisfaction
Low = normal issue

Return ONLY: High, Medium, or Low.
"""

# -----------------------------
# STEP 3: Response Generation Prompt
# -----------------------------

response_prompt = """
Generate a professional and empathetic response.
3–5 sentences.
Match tone to urgency.
Avoid blaming language.
"""

# -----------------------------
# STEP 4: Escalation Rule Prompt
# -----------------------------

escalation_prompt = """
If urgency is High:
Write an internal escalation note.
Otherwise:
Return 'No escalation required.'
"""
def classify_issue(message):
    message = message.lower()

    if "charge" in message or "refund" in message or "subscription" in message:
        return "Billing"
    elif "error" in message or "not working" in message or "bug" in message:
        return "Technical Support"
    elif "login" in message or "password" in message:
        return "Account Management"
    else:
        return "General Inquiry"


def detect_urgency(message):
    message = message.lower()

    if "cancel" in message or "legal" in message or "fraud" in message:
        return "High"
    elif "again" in message or "still" in message:
        return "Medium"
    else:
        return "Low"


def generate_response(message, category, urgency):

    if urgency == "High":
        return (
            "We sincerely apologize for the inconvenience you’ve experienced. "
            "We understand how frustrating this situation must be and appreciate you bringing it to our attention. "
            "Our team is prioritizing this matter and will work quickly to resolve it. "
            "Thank you for your patience while we address this issue."
        )

    elif urgency == "Medium":
        return (
            "Thank you for reaching out regarding this issue. "
            "We understand your concern and are reviewing your case carefully. "
            "Our team is working to resolve this as efficiently as possible."
        )

    else:
        return (
            "Thank you for contacting us. "
            "We’re happy to assist with your request and will ensure it is addressed promptly."
        )


def escalation_decision(urgency):
    if urgency == "High":
        return (
            "ESCALATION NOTE: Customer has expressed intent to cancel. "
            "Issue categorized as high urgency billing concern. "
            "Please prioritize immediate review."
        )
    else:
        return "No escalation required."
customer_message = "I was charged twice for my subscription and if this isn't fixed I will cancel immediately."

# Step 1
category = classify_issue(customer_message)

# Step 2
urgency = detect_urgency(customer_message)

# Step 3
response = generate_response(customer_message, category, urgency)

# Step 4
escalation = escalation_decision(urgency)

# Output Results
print("CUSTOMER MESSAGE:")
print(customer_message)

print("\nSTEP 1 — CATEGORY:")
print(category)

print("\nSTEP 2 — URGENCY:")
print(urgency)

print("\nSTEP 3 — CUSTOMER RESPONSE:")
print(response)

print("\nSTEP 4 — ESCALATION DECISION:")
print(escalation)
```

During initial testing, classification logic failed to detect cancellation threats.
I refined the urgency detection step by adding keyword triggers such as "cancel", "legal", and "fraud".

This improved escalation accuracy and ensured proper chaining between steps.
