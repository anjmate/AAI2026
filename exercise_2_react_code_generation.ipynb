You are a Python engineer using the ReACT methodology.

Task: Write a Python function that calculates the factorial of a number.

Follow this structure:

Stage 1 - Reason:
Explain your plan step-by-step.

Stage 2 - Act:
Write the Python code only.

Stage 3 - Observe:
Show what happens when factorial(5) is executed.

Stage 4 - Fix:
Identify potential edge cases (negative numbers, non-integers) and revise the code to handle them.


```python

print("STAGE 1 — REASON\n")

print("To calculate factorial:")
print("1. Factorial is the product of all positive integers up to n.")
print("2. 5! = 5 × 4 × 3 × 2 × 1.")
print("3. We can use a loop to multiply numbers from 1 to n.")
print("4. We must also consider edge cases like negative numbers or non-integers.\n")

print("STAGE 2 — ACT (Initial Version)\n")

def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print("Initial function created.")

print("\nSTAGE 3 — OBSERVE\n")

print("factorial(5) =", factorial(5))

print("\nTesting Edge Cases:\n")

try:
    print("factorial(-3) =", factorial(-3))
except Exception as e:
    print("Error with negative number:", e)

try:
    print("factorial(3.5) =", factorial(3.5))
except Exception as e:
    print("Error with float:", e)

print("\nSTAGE 4 — FIX (Improved Version)\n")

def factorial(n):
    # Type validation
    if not isinstance(n, int):
        raise TypeError("Input must be an integer.")

    # Negative validation
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers.")

    result = 1
    for i in range(1, n + 1):
        result *= i

    return result

print("Revised function created.")

print("\nRE-TESTING AFTER FIX:\n")

print("factorial(5) =", factorial(5))

try:
    print("factorial(-3) =", factorial(-3))
except Exception as e:
    print("Handled negative case correctly:", e)

try:
    print("factorial(3.5) =", factorial(3.5))
except Exception as e:
    print("Handled non-integer correctly:", e)
```

Initial implementation correctly calculated factorial for positive integers but failed to handle:

* Negative inputs

* Non-integer inputs

After observing incorrect behavior in edge case testing, input validation was added using:

* TypeError for non-integer inputs

* ValueError for negative numbers

This completed the ReACT cycle:
Reason → Act → Observe → Fix → Re-test
